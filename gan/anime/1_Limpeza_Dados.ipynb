```python
import os, shutil, json
import numpy as np
import pandas as pd
from PIL import Image
from pathlib import Path
from sklearn.model_selection import train_test_split
from keras.preprocessing.image import ImageDataGenerator

abs_path = r'%s' % os.getcwd().replace('\\','/')
```

# Análise dos dados

```python
images_folder = os.path.join(abs_path,'1_Mckinsey666').replace('\\','/')
images_DataFrame = pd.DataFrame(columns=['path','width','height','size'])
```

```python
tamanho_imagens = []
for arq in os.listdir(images_folder):
    try:
        image_path = os.path.join('1_Mckinsey666',arq).replace('\\','/')
        image = Image.open(image_path)
        width, height = image.size
        tamanho = str(width) + " " + str(height)
        imagem_df = pd.DataFrame({'path': [arq], 'width': [width], 'height': [height], 'size':[tamanho] })
        images_DataFrame = pd.concat([images_DataFrame, imagem_df])
        if tamanho not in tamanho_imagens:
            tamanho_imagens.append(tamanho)
    except:
        continue
tamanho_imagens.sort()
images_DataFrame = images_DataFrame.reset_index(drop=True)
```
