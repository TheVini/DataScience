**análise:** a coluna da variável de resposta, 'Attrition_Flag', não está devidamente balanceada entre as duas classes fornecidas ('Attrited Customer' e 'Existing Customer'). Portanto, pode ser necessário utilizar técnicas para remover ou criar elementos sintéticos devido ao desbalanceamento entre as classes. Entre as variáveis explicativas existem algumas variáveis muito desequilibradas/enviesadas (ex.: Credit_limit, Avg_Utilization_Ratio e etc.), como poder ser observado abaixo nos histogramas das variáveis explicativas quantitativas e nos BoxPlots mais abaixo das variáveis explicativas quantitativas.

```python
np.unique(dataset['Attrition_Flag'], return_counts=True)
```

```python
qtd_colunas_subplot = 2
primeiro_index = 1
qtd_linhas_subplot = math.ceil(len(dataset.iloc[:,primeiro_index:].columns)/qtd_colunas_subplot)

_, ax = plt.subplots(qtd_linhas_subplot, qtd_colunas_subplot, figsize=(25,100))

for count, column in enumerate(dataset.iloc[:,primeiro_index:].columns):
    current_ax = ax[count//qtd_colunas_subplot][count%qtd_colunas_subplot]
    current_ax.set_title(column)
    if dataset[column].dtype in ["int64", "float64"]:
        kde = True if dataset[column].dtype == "float64" else False
        sns.distplot(dataset[column], kde=kde, ax=current_ax)
    else:
        data = pd.DataFrame(dataset[column].groupby(dataset[column]).size().reset_index(name='Quantidade'))
        current_ax.set_ylabel("No of customers")
        # Ordenação decrescente baseada na quantidade de elementos em cada classe de cada coluna
        # Depois é necessário fazer reshape porque o método .values retorna uma matrix
        order = dataset[column].groupby(dataset[column]).size().reset_index(name='qtcd_cust').sort_values(by=['qtcd_cust'], ascending=False).iloc[:,:1].values
        if column != "Attrition_Flag":
            sns.countplot(x = dataset[column], edgecolor = 'black', saturation = 0.55, 
                          ax=current_ax, hue=dataset['Attrition_Flag'], order=order.reshape((order.shape[0])))
        else:
            sns.countplot(x = dataset[column], edgecolor = 'black', saturation = 0.55, 
                          ax=current_ax, order=order.reshape((order.shape[0])))
        current_ax.set_ylabel('No of customers')
        
# Deletar os frames vazios
qtd_frames_vazios = qtd_colunas_subplot*qtd_linhas_subplot - len(dataset.iloc[:,primeiro_index:].columns)
[ax.flat[-i].set_visible(False) for i in range(1, qtd_frames_vazios+1)]
plt.show()
plt.close()
```

```python
colunas_numericas = (dataset.iloc[:,1:].dtypes != "object").reset_index(name='Quantidade').query('Quantidade == True')['index'].values

qtd_colunas_subplot = 4
qtd_linhas_subplot = math.ceil(len(colunas_numericas)/qtd_colunas_subplot)
_, ax = plt.subplots(qtd_linhas_subplot, qtd_colunas_subplot, figsize=(25,50))

for count, column in enumerate(colunas_numericas):
    current_ax = ax[count//qtd_colunas_subplot][count%qtd_colunas_subplot]
    current_ax.set_title(column)
    plt.subplot(sns.boxplot(data=dataset[column], ax=current_ax))

# Deletar os frames vazios
qtd_frames_vazios = qtd_colunas_subplot*qtd_linhas_subplot - len(colunas_numericas)
[ax.flat[-i].set_visible(False) for i in range(1, qtd_frames_vazios+1)]
plt.show()
plt.close()
```

```python
# Encodar os dados

dataset_encoded = dataset.iloc[:,1:].copy()
for column in dataset_encoded.columns:
    if dataset_encoded[column].dtype == "object":
        label_encoder = LabelEncoder()
        dataset_encoded[column] = label_encoder.fit_transform(dataset_encoded[column])
```

**Análise:** Para a tabela de correlação de Pearson mostrada através do mapa de calor abaixo, fica evidente que nenhuma das variáveis explicativas é no mínimo moderada, segundo a classificação de Evans (1996, ver também http://leg.ufpr.br/~silvia/CE003/node74.html)

```python
fig, ax = plt.subplots(figsize = (20, 20))
plt.title("Correlação de pares das colunas", fontsize = 18)
ax = sns.heatmap(dataset_encoded.iloc[:,:].corr(), annot = True)
```

**Análise:** Através dos gráficos de dispersão e distribuições abaixo, fica evidente que será bem difícil para os algoritmos conseguirem fazer distinções entre as duas classes da vairável "Attrition_Flag" dado as variáveis explicativas.

```python
vars = ['Total_Trans_Ct', 'Total_Ct_Chng_Q4_Q1', 'Total_Revolving_Bal', 'Avg_Utilization_Ratio', 
        'Total_Trans_Amt']
sns.pairplot(dataset, vars=vars, hue = "Attrition_Flag")
plt.show()
```

**Análise:** transformação logarítmica nos dados desbalanceados

```python
dataset_encoded_transformed = dataset_encoded.copy()
```

```python
tranform_col_list = ['Credit_Limit','Avg_Open_To_Buy','Total_Amt_Chng_Q4_Q1',
                     'Total_Ct_Chng_Q4_Q1','Avg_Utilization_Ratio','Total_Trans_Amt', 'Total_Revolving_Bal']
for column in tranform_col_list:
    _, ax = plt.subplots(1, 2, figsize=(20,10))
    first_col = ax[0]
    first_col.set_title(column + " - Before log transformation")
    sns.distplot(dataset_encoded_transformed[column], kde=True, ax=first_col)
    second_col = ax[1]
    second_col.set_title(column + " - After log transformation")
    sns.distplot(np.log1p(dataset_encoded_transformed[column]), kde=True, ax=second_col)
```
